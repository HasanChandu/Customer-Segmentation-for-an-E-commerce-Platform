```python
import pandas as pd

# Load your dataset
# Replace 'your_dataset.csv' with the path to your dataset
df = pd.read_csv('ecom_data.csv')

# Define a function to detect outliers using IQR
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

# Apply the function to detect outliers for each numerical column
outliers = df.select_dtypes(include=['number']).apply(detect_outliers_iqr)

# Display columns with outliers
print("Columns with outliers:")
print(outliers.any())  # This will print True for columns containing outliers
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
# Replace 'your_dataset.csv' with the path to your dataset
df = pd.read_csv('ecom_data.csv')

# Pairplot to visualize relationships between variables
sns.pairplot(df)
plt.title("Pairplot of Variables")
plt.show()

# Correlation matrix
correlation_matrix = df.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# Specific relationship visualization (e.g., scatter plot)
sns.scatterplot(x='column1', y='column2', data=df)
plt.title("Scatter Plot of column1 vs column2")
plt.show()

# Additional visualizations such as histograms, box plots, etc., can be added as needed
```
